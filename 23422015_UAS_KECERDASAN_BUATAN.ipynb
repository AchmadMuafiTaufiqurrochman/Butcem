```python
# Mendefinisikan hubungan orangtua
orangtua = {
    "Hadi": ["Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian"],
    "Desi": ["Wanda", "Aji", "Gunawan"],
    "Bayu": ["Tari", "Nurul", "Yanto"],
    "Retno": ["Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian"],
    "Rina": ["Hamzah", "Eka", "Mira", "Bastian"],
    "Ardi": ["Anggun", "Boy"],
    "Muafi": ["Eka", "Mira", "Bastian", "Anggun", "Boy"],
    "Fahrul": ["Aji", "Gunawan", "Anggun", "Boy"],
}

# Mendefinisikan hubungan saudara laki-laki
saudaralaki = {
    "Wahyu": ["Ardi", "Rina", "Desi"],
    "Ardi": ["Wahyu", "Rina", "Desi"],
    "Eka": ["Bastian", "Mira"],
    "Bastian": ["Eka", "Mira"],
    "Yanto": ["Nurul", "Yanti"],
    "Aji": ["Gunawan"],
    "Gunawan": ["Aji"],
    "Boy": ["Anggun"],
}

# Mendefinisikan hubungan saudara perempuan
saudaraperempuan = {
    "Rina": ["Ardi", "Wahyu", "Desi"],
    "Desi": ["Ardi", "Wahyu", "Rina"],
    "Eka": ["Mira", "Bastian"],
    "Mira": ["Eka", "Bastian"],
    "Tari": ["Nurul", "Yanto"],
    "Nurul": ["Tari", "Yanto"],
    "Anggun": ["Boy"],
}

# Mendefinisikan hubungan paman
paman = {
    "Bayu": ["Hamzah", "Eka", "Mira", "Bastian"],
    "Wahyu": ["Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian"],
    "Ardi": ["Tari", "Nurul", "Yanto", "Hamzah"],
    "Fahrul": ["Aji", "Gunawan", "Anggun", "Boy"],
    "Yanto": ["Wanda", "Aji", "Gunawan", "Anggun", "Boy"],
    "Hamzah": ["Wanda", "Aji", "Gunawan", "Anggun", "Boy"],
    "Eka": ["Gunawan", "Aji", "Wanda"],
    "Bastian": ["Gunawan", "Aji", "Wanda"],
}

# Mendefinisikan hubungan kakek
kakek = {
    "Hadi": ["Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian"],
    "Bayu": ["Wanda", "Aji", "Gunawan"],
    "Ardi": ["Anggun", "Boy"],
    "Muafi": ["Anggun", "Boy"],
}

# Mendefinisikan hubungan nenek
nenek = {
    "Retno": ["Tari", "Nurul", "Yanto", "Hamzah", "Eka", "Mira", "Bastian"],
    "Desi": ["Wanda", "Aji", "Gunawan"],
    "Ana": ["Anggun", "Boy"],
}

# Mendefinisikan hubungan sepupu
sepupu = {
    "Tari": ["Hamzah", "Eka", "Mira", "Bastian"],
    "Nurul": ["Hamzah", "Eka", "Mira", "Bastian"],
    "Yanto": ["Hamzah", "Eka", "Mira", "Bastian"],
    "Hamzah": ["Eka", "Mira", "Bastian", "Tari", "Nurul", "Yanto"],
    "Wanda": ["Aji", "Gunawan", "Anggun", "Boy"],
    "Aji": ["Wanda", "Anggun", "Boy"],
    "Gunawan": ["Wanda", "Anggun", "Boy"],
    "Anggun": ["Gunawan", "Aji", "Wanda"],
    "Boy": ["Gunawan", "Aji", "Wanda"],
}

# Fungsi untuk menampilkan hubungan dengan format yang diinginkan
def tampilkan_hubungan(hubungan, kategori):
    for key, value in hubungan.items():
        print(f"{key} adalah {kategori} dari : {', '.join(value)}")

# Menampilkan hubungan dengan format yang diinginkan
print("Orangtua:")
tampilkan_hubungan(orangtua, "orang tua dari")
print("\nSaudara Laki-laki:")
tampilkan_hubungan(saudaralaki, "saudara laki-laki dari")
print("\nSaudara Perempuan:")
tampilkan_hubungan(saudaraperempuan, "saudara perempuan dari")
print("\nPaman:")
tampilkan_hubungan(paman, "paman dari")
print("\nKakek:")
tampilkan_hubungan(kakek, "kakek dari")
print("\nNenek:")
tampilkan_hubungan(nenek, "nenek dari")
print("\nSepupu:")
tampilkan_hubungan(sepupu, "sepupu dari")
```
